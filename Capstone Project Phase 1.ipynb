# 1. Trying to collect the population of each state.

**While trying to collect the population data I face a problem like here we can get only 2001 and 2011's population data.So I will scrap these data then I will check the population growth for each state and I will fill the rest of the year's data according to that.I will mention from where I got the 2001 and 2011 population data and from where I got the growth rate in below.Please check those:**

https://m.rbi.org.in/Scripts/PublicationsView.aspx?id=22065

**For growth rate I took the data from below link:**

# Scraping the growth rate for each state

```python
#Will open crome
driver=webdriver.Chrome()
#Will access the link on that crome tab
driver.get("https://m.rbi.org.in/Scripts/PublicationsView.aspx?id=16988")
```

```python
GR=[]
growth_rate=driver.find_elements(By.XPATH,'//td')
for i in growth_rate:
    growth=i.text
    GR.append(growth)
```

```python
GR[18::7]
```

```python
raw_data=pd.DataFrame({"States/Union Territories":everything[13::8],"Growth Rate(2001-2011)":GR[18::7],"2001":everything[19::8],"2011":everything[20::8]})
raw_data
```

**So now I am going to create 2002 to 2010 population data based on the growth rate and I will use a formula to increase the population.**

```python
estimated_population_list = []

for index, state_data in raw_data.iterrows():
    growth_rate = float(state_data["Growth Rate(2001-2011)"])
    initial_population_2001 = float(state_data["2001"])  # Convert to numeric type

    # Calculating the annual growth 
    annual_growth = 1 + (growth_rate / 100) / 9

    
    estimated_population = {"States/Union Territories": state_data["States/Union Territories"]}
    for year in range(2002, 2011):
        estimated_population[str(year)] = round(initial_population_2001 * (annual_growth ** (year - 2001)))

    
    estimated_population_list.append(estimated_population)


estimated_population_df = pd.DataFrame(estimated_population_list)


estimated_population_df
```

```python
new_population=pd.merge(raw_data,estimated_population_df,on="States/Union Territories")
new_population
```

```python
#Changing the 2011 columns location and set it to the last
new_population= new_population[['States/Union Territories', 'Growth Rate(2001-2011)', '2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011'] ]
new_population
```

**Observation:-** 
    
*Here you can see 2011 data is correct because I scrap it from the RBI site but if you look closely here you can see for 2010 population it's higher that the 2011 and that should not be possible so I will reduce the population of 2010 for 1.5%.Lets see:*

```python
#Decreasing the 2010 columns values by 1.5 %
new_population['2010'] = round(new_population['2010'] * (1 - 1.50 / 100))
```

```python
new_population
```

```python
#Removing the decimal part of the 2010 columns data
new_population['2010']=new_population['2010'].astype(int)
```

```python
new_population
```

**Here you can see decimal part has been removed for the 2010 column.**

**Now I will add 2012 column and for that I will increase 2011 columns values to 1.7% because from 2001 to 2011 growth rate was 17.64 and if I divide it with 10 then the result is each year 1.764.Thats why I will take 1.7.** 

```python
#Changing the datatype of 2011 column first
new_population['2011']=pd.to_numeric(new_population['2011'])
```

```python
#Adding 2012 column and data in it should be the 2011's data with 1.7% increament
new_population['2012']=round(new_population['2011'] * (1+1.7/100)).astype(int)
```

```python
new_population
```

**So here I don't need the Growth Rate(2001-2011) column so I am deleting this column.**

```python
new_population.drop("Growth Rate(2001-2011)",axis=1,inplace=True)
new_population
```

**Now you can see Growth Rate(2001-2011) column is deleted.**

```python
#Replaceing Jammu & Kashmir name with Jammu and Kashmir so I can merge easily in the later part
new_population["States/Union Territories"].replace("Jammu & Kashmir","Jammu and Kashmir",inplace=True)
```

# Creating another shape of Population dataset

```python
#Reshaping the dataset according the file I got
population_df = new_population.melt(id_vars=['States/Union Territories'], var_name='Year', value_name='Population')

# Converting the 'Year' column to numeric
population_df['Year'] = population_df['Year'].astype(int)

# Sort the DataFrame for better presentation
population_df.sort_values(['States/Union Territories', 'Year'], inplace=True)


population_df
```

# 2. Scraping the literacy Rate in each state

**For the each state literacy rate I took the data from below link:**

https://www.rbi.org.in/scripts/PublicationsView.aspx?id=19995

```python
#Will open crome
driver=webdriver.Chrome()
#Will access the link on that crome tab
driver.get("https://www.rbi.org.in/scripts/PublicationsView.aspx?id=19995")
```

```python
literacy=[]

lit_data=driver.find_elements(By.XPATH,'//td')
for i in lit_data:
    lit_text=i.text
    literacy.append(lit_text)
    if len(literacy)==301:
        break
```

```python
Literacy=pd.DataFrame({"States/Union Territories":literacy[13::8],"2001":literacy[19::8],"2011":literacy[20::8]})
Literacy
```

**Observation:-**
    
*1) Here you can see the literacy rate of only 2001 and 2011 but I have the data of 2001 to 2012 so for rest of the columns I have to fill that.*

*2) From 2001 to 2011 ,in between there are 9 years and in 2001 the literacy rate of of India is 64.84 and of 2011 that is 72.99 meas it increases like 8.15% in these 9 years that means 0.90% increament each year so I will create the 2002 to 2010 and 2012 literacy rate data by increamenting the values for each year.* 

```python
#Changing the datatype of 2001 column first
Literacy['2001']=pd.to_numeric(Literacy['2001'])
```

```python
#Adding 2002 column and data in it should be the 2001's data with 0.37% increament
Literacy['2002']=round(Literacy['2001'] * (1+0.37/100),2).astype(float)
```

```python
#Now I will create different columns of year with incremented literacy rate
Literacy['2003']=round(Literacy['2002'] * (1+0.45/100),2).astype(float)
Literacy['2004']=round(Literacy['2003'] * (1+0.37/100),2).astype(float)
Literacy['2005']=round(Literacy['2004'] * (1+0.33/100),2).astype(float)
Literacy['2006']=round(Literacy['2005'] * (1+0.26/100),2).astype(float)
Literacy['2007']=round(Literacy['2006'] * (1+0.52/100),2).astype(float)
```

```python
#Adding more columns
Literacy['2008']=round(Literacy['2007'] * (1+0.59/100),2).astype(float)
Literacy['2009']=round(Literacy['2008'] * (1+1.00/100),2).astype(float)
Literacy['2010']=round(Literacy['2009'] * (1+0.56/100),2).astype(float)
#Checging the datatype of Literacy['Literacy Rate of 2011'] column
Literacy['2011']=pd.to_numeric(Literacy['2011'])
Literacy['2012']=round(Literacy['2011'] * (1+0.45/100),2).astype(float)
```

```python
#setting the column in ascending order
Literacy=Literacy[["States/Union Territories","2001","2002","2003","2004","2005","2006","2007","2008","2009","2010","2011","2012"]]
Literacy
```

**Observation:-**
    
1) Now you can see my literacy dataset is ready and it has 2001 to 2012 data in it.
