# 2.1 Analysis of Literacy Rate vs Total Crimes.

**Here I will analize different sections of crime's total crime with Literacy rate.**

```python
#Creating a IPC crime dataset state and year wise
total_IPC_crimes_Statewise = crime_IPC.groupby(['STATE/UT', 'YEAR'])['TOTAL IPC CRIMES'].sum().reset_index()
total_IPC_crimes_Statewise
```

```python
Literacyrate_Statewise=Combine[["STATE/UT","YEAR","Literacy Rate"]]
Literacyrate_Statewise
```

```python
Literacyrate_Statewise["STATE/UT"].value_counts()
```

```python
#Creating a new Literacy dataset without ALL INDIA value
Literacyrate_Statewise_without_ALL_INDIA_data = Literacyrate_Statewise[
    Literacyrate_Statewise["STATE/UT"] != "ALL INDIA"
]
Literacyrate_Statewise_without_ALL_INDIA_data
```

```python
#Exporting Literacy data
Literacyrate_Statewise_without_ALL_INDIA_data.to_csv("Lireracy_Data(2001-2012).csv",index=False)
```

```python
Total_IPC_crimes_state_wise_with_Literacy=pd.merge(total_IPC_crimes_Statewise,Literacyrate_Statewise,on=["STATE/UT","YEAR"])
Total_IPC_crimes_state_wise_with_Literacy
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Literacy Rate', y='TOTAL IPC CRIMES', data=Total_IPC_crimes_state_wise_with_Literacy)
plt.title('Literacy Rate vs Total IPC Crimes (2001-2012)')
plt.xlabel('Literacy Rate')
plt.ylabel('Total IPC Crimes')
plt.show()
```

**Observation:-**
    
*1) We can't get much of the details but you can see where literacy rate is between 60 to 80 there IPC crime is very high.*

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='YEAR', y='Literacy Rate', data=Total_IPC_crimes_state_wise_with_Literacy, label='Literacy Rate')
sns.lineplot(x='YEAR', y='TOTAL IPC CRIMES', data=Total_IPC_crimes_state_wise_with_Literacy, label='Total IPC Crimes')
plt.title('Literacy Rate and Total IPC Crimes (State-wise, 2001-2012)')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.show()
```

**Observation:**

*1) Here you can see Literacy rate is seemed not increasing or nither decreasing but Crime rate is increasing year by year.*

```python
Correlation_IPC_And_Literacy=Total_IPC_crimes_state_wise_with_Literacy[['Literacy Rate', 'TOTAL IPC CRIMES']].corr()
sns.heatmap(Correlation_IPC_And_Literacy,cmap="Paired",fmt="0.2f",linewidth=0.2,annot=True,linecolor="black")
plt.xlabel("Features Name",fontsize=14)
plt.title("Correlation Graph",fontsize=20)
plt.show()
```

**Observation:-**
    
*Here you can see TOTAL IPC CRIMES and Literacy Rate has a negative correlation but that is not too strong.*
